# **Road Data Information Visualization using python**





```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading the database
data = pd.read_csv("Road_Transport_BRS_2017-18.csv")
 
# printing the data
display(data)
```

```python
# Setting a custom figure size
fig = plt.figure(figsize =(20, 10))

# Horizontal Bar Chart
plt.barh(data['Name of the States'], data['Total road Length'])

plt.title("Bar Chart")

# Setting the X and Y labels
plt.ylabel('Name of the States')
plt.xlabel('Total Road Length(km)')

plt.show()
```

```python
# Setting a custom figure size
fig = plt.figure(figsize =(30, 10))

# Setting the X and Y labels
plt.xlabel('Name of the States')
plt.ylabel('Length in km')

# List of columns
l=['National Highways',	'State Highways',	'District Roads',	'Rural Roads', 'Urban roads',	'Project Roads']

# Line Graph
for i in l:
  plt.plot(data['Name of the States'],data[i],label=i)

plt.legend()
plt.show()
```

```python
# Extracting Road Density Data
y=np.array(data['Road Density '])

# Pie Chart
plt.pie(y,labels=data['Name of the States'])

plt.show()
```

```python
# Converting Road Density into per sq.km
x=(data['Road Density ']/1000)**2

# Calculating Population Density per sq.km
z=data['total Population']/data['Total Area']

fig,ax = plt.subplots()

# Road Density bar graph
ax.bar(data['Name of the States'], z)

# Setting the X and Y labels
ax.set_xlabel('Name of the States')
ax.set_ylabel('Road Density (per sq.km)')

ax2=ax.twinx()

# Population Desnity line graph
ax2.plot(data['Name of the States'],x,color='green')

# Setting the Y label
ax2.set_ylabel('Population Density (per sq.km)')

plt.show()
```
